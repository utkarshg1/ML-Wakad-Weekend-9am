# Create a function to replace missing values

```python
X.columns
```

```python
X.dtypes
```

```python
cat_cols = list(X.columns[X.dtypes == "object"])
cat_cols
```

```python
con_cols = list(X.columns[X.dtypes != "object"])
con_cols
```

```python
df["Price"].mean()
```

```python
df["Price"].median()
```

```python
def replacer(df: pd.DataFrame):
    cat_cols = list(df.columns[df.dtypes == "object"])
    
    for col in df.columns:
        
        if col in cat_cols:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
        
        else:
            mean = df[col].mean()
            df[col] = df[col].fillna(mean)

    print(f"Missing Values replaced in dataframe")
```

```python
replacer(X)
```

```python
X.isna().sum()
```

### Data Cleaning complete

### Feature Engineering

```python
X.head()
```

```python
X_cat = X[cat_cols]
X_cat
```

```python
X_con = X[con_cols]
X_con
```

# Categorical Columns apply one hot encoding

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False, drop="first"
).set_output(transform = "pandas")

X_cat_pre = ohe.fit_transform(X_cat)

X_cat_pre.head()
```

# Continuous features Apply StandardScaler

```python
X_con
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform = "pandas")
X_con_pre = scaler.fit_transform(X_con)
X_con_pre.head()
```

# Combine categorical and continuous features

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

# Preprocessing complete

# Step 5 - Train Test Split

### test size can be from 0.2 to 0.4

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, test_size=0.2, random_state=21
)
```

# Step 6 - Build the model

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print("Train Metrics : ")
evaluate_model(model, xtrain, ytrain)
```

```python
print("Test Metrics : ")
evaluate_model(model, xtest, ytest)
```
